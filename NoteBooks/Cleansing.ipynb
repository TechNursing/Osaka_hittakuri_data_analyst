```python
import pandas as pd

hittak = pd.read_csv('../raw/osaka_2021hittakuri.csv')
display(hittak.head(5))
```

```python
hittak['発生年月日（始期）'] = hittak['発生年月日（始期）'].str.split('-')
hittak['発生年月日（始期）'] = hittak['発生年月日（始期）'].str[1]
hittak['発生年月日（始期）'] = hittak['発生年月日（始期）'].astype(int)
print(f'{hittak.dtypes} \n')
display(hittak.head(5))
```

```python
hittak_map = hittak[['被害者の年齢','被害者の性別','市区町村（発生地）','発生時（始期）', '発生年月日（始期）']]
display(hittak_map.head(5))
```

```python
hittak_map = hittak_map.sort_values(by=['発生時（始期）'])
display(hittak_map.head(5))
display(hittak_map.tail(5))
```

```python
hittak_map.to_csv('../CleansedFiles/hittak_map.csv', mode='w', index=False)
```
