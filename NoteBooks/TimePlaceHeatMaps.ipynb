```python
# ヒートマップで表すために、数値型になっている時間をカテゴリ化させる
hittak_map['発生時（始期）'] = hittak_map['発生時（始期）'].astype(str)

# 各columnのデータ型を表示
print(f'{hittak_map.dtypes} \n')
```
